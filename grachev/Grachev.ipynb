```python
canv = ROOT.TCanvas("canv","canv",900,450)
```

```python
h = ROOT.TH1F("h",";x;events",20,-20,20)
with open("out_20.data","r") as fl:
    for line in fl:
        w = line.split(" ")
        if float(w[5])!=0:
            h.Fill(float(w[8]))
                   
h.Draw()
ROOT.gPad.SetLogy()
canv.Draw()
```

```python
h5 = ROOT.TH1F("h5",";x;events",20,-20,20)
h5.SetLineColor(2)
h5.SetFillColor(2)
h5.SetFillStyle(3005)
with open("out_05.data","r") as fl:
    for line in fl:
        w = line.split(" ")
        if float(w[5])!=0:
            h5.Fill(float(w[8]))
h5.Draw()
h.Draw("same")
ROOT.gPad.SetLogy()
canv.Draw()
```

```python
h.scale(1.)
h5.scale(1)
h5.Draw()
h.Draw("same")
ROOT.gPad.SetLogy()
canv.Draw()
```
